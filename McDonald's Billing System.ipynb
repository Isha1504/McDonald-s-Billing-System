# Calculator

```python
cal_text=""
text=StringVar()

def Number(x):
    global cal_text
    cal_text = cal_text+str(x)
    text.set(cal_text)
def Clear():
    global cal_text
    cal_text = ""
    text.set("")
def Calculate():
    global cal_text
    result=str(eval(cal_text))
    text.set(result)
    cal_text=""
```
